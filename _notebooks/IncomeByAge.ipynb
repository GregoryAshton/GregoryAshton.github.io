```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import wget

# Dependencies: wget, xlrd
```

```python
file_names = [  
"table3_2_september04.xls",
"table3-2-2000-01.xls",
"table-32-2001-02.xls",
"table3_2.xls",
"3_2_apr06.xls",
"table3-2-2004-05.xls",
"table3-2-jan08.xls",
"3-2tabledec08.xls",
"3-2table-jan2010.xls",
"3-2table-feb2012.xls"]

years = ["{}-{}".format(i, i+1) for i in range(1999, 2010)]
```

```python
url_base = "http://webarchive.nationalarchives.gov.uk/20120405152450/http://hmrc.gov.uk/stats/income_distribution/{}"


for file, year in zip(file_names, years):
    fname = wget.download(url_base.format(file))
    print "Downloaded {} for year {}".format(fname, year) 
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

for age in np.unique(df.age.values):
    ax.plot(df[df.age == age]['MeanIncomeBeforetax'], "-o", lw=2, label=age)
    ax.set_xticklabels(years, rotation=45)

plt.legend(framealpha=0.2)
plt.show()
```
