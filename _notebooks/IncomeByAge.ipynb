## Combing all data sets

```python
df_list = [df_1999_2000, df_2000_2001, df_2001_2002,
           df_2002_2003, df_2003_2004, df_2004_2005,
           df_2005_2006, df_2006_2007, df_2007_2008,
           df_2008_2009, df_2009_2010, df_2010_2011, 
           df_2011_2012, df_2012_2013]

years_total = np.concatenate((years_1999_2009, years_2010_2013))
df = pd.concat(df_list, keys=years_total) 

df
```

## Plotting the data

```python
years_total_val = [int(s.split("-")[0]) for s in years_total]

age_ranges = df.ix['1999-2000'].age.values

fig, ax = plt.subplots(figsize=(10, 5))
NUM_COLORS = len(age_ranges)
cm = plt.get_cmap('gist_rainbow')
ax.set_color_cycle([cm(1.*i/NUM_COLORS) for i in range(NUM_COLORS)])

for age in age_ranges:
    ax.plot(years_total_val, df[df.age == age]['MeanIncomeBeforetax'], "-o", 
            lw=2, label=age)
    ax.set_xticks(years_total_val)
    ax.set_xticklabels(years_total, rotation=45)

plt.legend(bbox_to_anchor=(1.3, 1.0))
plt.show()
```

## Distribution

```python
age_ranges_values = range(len(age_ranges))

fig, ax = plt.subplots(figsize=(10, 5))
NUM_COLORS = len(years_total)
cm = plt.get_cmap('gist_rainbow')
ax.set_color_cycle([cm(1.*i/NUM_COLORS) for i in range(NUM_COLORS)])

for yr in years_total:
    ax.plot(age_ranges_values, df.ix[yr]['MeanIncomeBeforetax'], "-o",
            label=yr)
    
ax.set_xticks(age_ranges_values)
ax.set_xticklabels(age_ranges, rotation=45)
plt.legend(bbox_to_anchor=(1.3, 1.0))
plt.show()
```
